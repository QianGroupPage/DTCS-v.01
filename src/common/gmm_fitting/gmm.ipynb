```python
from gmm import *
from util import *

import matplotlib.pyplot as plt
import numpy as np
import cProfile

%matplotlib inline
%load_ext autoreload
```

```python
# TODO Wrap into expiriment program w/ argparser
n = 5
xpsModel = XPSModel(n=n)

n_expiriment_cols = 3
n_expiriment_rows = 5
figure_width = 25
fig, ax = plt.subplots(n_expiriment_rows, n_expiriment_cols, sharex='col', sharey='row', figsize=(figure_width, figure_width / n_expiriment_cols * n_expiriment_rows / 3))
fig.subplots_adjust(hspace=0, wspace=0)

for col in range(n_expiriment_cols):
    for row in range(n_expiriment_rows):
        targetModel = xpsModel.sample()
        targetModel.plot(ax[row, col], data_fmt='b.', species_fmt='b-', verbose=False, data=True)
        try:
            learnedModel = xpsModel.fit(targetModel, guess_n=n+2)
            learnedModel.plot(ax[row, col], data_fmt='r,', species_fmt='r--', verbose=False, data=True)
        except RuntimeError as e:
            pass
# fig.legend()
```
