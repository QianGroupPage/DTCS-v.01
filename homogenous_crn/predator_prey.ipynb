```python
import crn_sym as crn
import solution as sol

x1, x2 = crn.species("x1 x2")
```

```python
rsys = crn.RxnSystem(
   crn.Rxn(x1 + x2, 2 * x2, 1.5),
   crn.Rxn(x1, 2 * x1, 1),
   crn.Rxn(x2, 1, 1),
   
   crn.ConcEq(x1, 2),
   crn.ConcEq(x2, 1)
)

tmax = 45
```

```python
solution = sol.solve(rsys, time = tmax)

sol = solution.basic_plot()
```
