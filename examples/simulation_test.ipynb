```python
from lblcrn.homogenous_crn.sym import *
import matplotlib.pyplot as plt

from math import sin
```

```python
sm = SpeciesManager()

x1 = sm.sp('x', [Orbital('1s', 530), Orbital('2s', 532)], {0:2})
x2 = sm.sp('y', Orbital('1s', 531), {0:1})
#x3 = sm.sp('z', Orbital('1s', 611), 20, {0:1})

rsys = RxnSystem(
    Rxn(x1+ x2, 2*x2, 1.5),
    Rxn(x1, 2 * x1, 1),
    Rxn(x2, 1, 1),
    sm
)

```

```python
s = solve(rsys, time=40, max_step=1e-2)
```
