```python
from lblcrn.crn_sym import *
from lblcrn.experiments.xps import simulate_xps
from lblcrn.experiments.time_series import simulate_crn
```

```python
sm = SpeciesManager()
prey = sm.sp('rabbit', Orbital('1s', 1.0))
pred = sm.sp('fox', Orbital('1s', 2))

rsys = RxnSystem(
    sm,

    Rxn(prey + pred, 2 * pred, 2),
    Rxn(prey, 2 * prey, 1),
    Rxn(pred, 1, 1),

    Conc(pred, 1),
    Conc(prey, 1),
)
```

```python
s = simulate_crn(rsys, time=45, max_step=0.1)
```

```python
s.plot()
```
