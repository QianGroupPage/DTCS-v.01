```python
sm = SpeciesManager()

x1 = sm.sp('X #1', Orbital('1s', 544), 30, {0:1})
x2 = sm.sp('X #2', Orbital('1s', 532), 30, {0:1})
```

```python
rsys = RxnSystem(
    Rxn(x1 + x2, 2 * x2, 1.5),
    Rxn(x1, 2 * x1, 1),
    Rxn(x2, 1, 1),
    
    Conc(x1, 2),
    Conc(x2, 1),
    sm
)

solution = solve(rsys, time=45, max_step=0.1)
```
